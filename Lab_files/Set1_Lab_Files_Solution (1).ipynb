```python
print("Temperature Converter")
print("---------------------")

choice = input("Choose an option:\n1. Fahrenheit to Celsius\n2. Celsius to Fahrenheit\nEnter 1 or 2: ")

if choice == '1':
    f = float(input("Enter temperature in Fahrenheit: "))
    c = (f - 32) * 5/9 #c==5.89897867
    print(f"{f} degrees Fahrenheit is equal to {c:.5f} degrees Celsius.")

elif choice == '2':
    c = float(input("Enter temperature in Celsius: "))
    f = (c * 9/5) + 32
    print(f"{c} degrees Celsius is equal to {f:.4f} degrees Fahrenheit.")

else:
    print("Invalid choice. Please enter 1 or 2.")
```

**A program that calculates the area and perimeter of a rectangle.**

```python
print("Rectangle Calculator")
print("---------------------")

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area = length * width
perimeter = 2 * (length + width)

print(f"The area of the rectangle is: {area:.2f} square units")
print(f"The perimeter of the rectangle is: {perimeter:.2f} units")
```

# **3. A program that generates a random password of a specified length.**




```python
import random
import string

print("Random Password Generator")
print("-------------------------")

password_length = int(input("Enter the desired length for the password: "))

if password_length > 0:
    password =''.join(random.choice(string.ascii_letters + string.digits + string.punctuation) for _ in range(password_length))
    print(f"Generated Password: {password}")
else:
    print("Invalid password length. Please enter a positive integer.")
```

# **4. A program that calculates the average of a list of numbers**

```python
print("Average Calculator")
print("-------------------")

numbers = input("Enter a list of numbers separated by spaces: ")
numbers_list = [float(num) for num in numbers.split()]#Enter a list of numbers separated by spaces: 34 4
                                                               #[34.0, 4.0]
if numbers_list:
    average = sum(numbers_list) / len(numbers_list)
    print(f"The average is: {average:.2f}")
else:
    print("No numbers provided. Please enter a list of numbers.")
```

##5. A program that checks if a given year is a leap year.
image.png

```python
# Get user input for the year
year = int(input("Enter a year: "))

# Check if it's a leap year
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
      print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

#6. A program that calculates the factorial of a number.

image.png

```python
#factorial
def factorial(n):
    """This is a recursive function to find the factorial of an integer"""

    if n == 1:
        return 1
    else:
        return (n * factorial(n-1))  # 3 * 2 * 1 = 6

num = 3
print("The factorial of", num, "is", factorial(num))
```

# 8. A program that sorts a list of numbers in ascending or descending order.

```python
# Get user input for the list of numbers
numbers = [1,4,2,0,7,9,4,6,2,9]

#numbers = [float(num) for num in input("Enter a list of numbers separated by spaces: ").split()]

# Get user input for sorting order
sort_order = input("Enter 'asc' for ascending order or 'desc' for descending order: ")

# Sort and print the list based on user input
sorted_numbers = sorted(numbers, reverse=(sort_order.lower() == 'desc'))
print(f"Sorted list in  {sorted_numbers}")
```

# 9. A program that generates a multiplication table for a given number.

```python
# Get user input for the number
number = int(input("Enter a number: "))

# Generate and print the multiplication table
print(f"Multiplication table for {number}:")
for i in range(1, 11):
    result = number * i
    print(f"{number} x {i} = {result}")
```
